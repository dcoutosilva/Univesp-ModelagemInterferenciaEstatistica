```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
df = pd.read_csv(io.StringIO(uploaded['cars_vertical1.txt'].decode('utf-8')))
df.head(15)
```

```python
#exibindo as informações
df.info()
#computa estatistica
df.describe()
```

```python
#tratar os dados
c0=df["brand/model/year"]
c1=df["cap_vol"]
c2=df["consumo"]
c3=df["power"]
c4=df["weight"]
c5=df['cemm']
c6=df["nu_cy"]
c7=df["Etype"]
```

```python
import stemgraphic as stg
# Observar todas as variáveis e escolher as adequadas
# c1, scale=1000 --> Capacidade volumétrica ou cilindrada
# c2, scale=1    --> Consumo
# c3, scale=100  --> Potência
# c4, scale=100  --> Peso
# c5, scale=10   --> 0-100

stg.stem_graphic(c1, scale=1000)
stg.stem_graphic(c2, scale=1)
```

```python
#exercicio 2 - fazer o grafico comparando consumo e Cilindradas
#ordenado = np.argsort(c0)

plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':100})
plt.scatter(c1,c2, label=f'Correlação = {np.round(np.corrcoef(c1,c2)[0,1],2)}')
plt.xlabel('Capacidade volumétrica')
plt.ylabel('consumo de combustível')
#plt.xlabel(f'{c0}')
#plt.tick_params()
#plt.xticks(ordenado, c0[ordenado])
plt.title('Gráfico de dispersão: Consumo x Cilindradas')
plt.legend()
plt.grid(True)
plt.show()
```
